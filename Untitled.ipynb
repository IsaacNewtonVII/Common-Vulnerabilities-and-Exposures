# OBJECTIVES AND GOALS 

Our goal is to predict the Common Vulnerability Scoring System(CVSS) score, in other words we want to predict the severity of the vulnerability in our system.
We also want to know the effect of each of our feature on our CVSS score. In the next session, we analyse our data and bring out useful findings from our dataset.

```python
#Read our products dataset
products = pd.read_csv('products.csv')
products.rename(columns={'cve_id': 'Unnamed: 0'}, inplace = True)
products.head()
```

```python
products.info()
```

```python
#Read in our vendor product dataset
ven_prod = pd.read_csv('vendor_product.csv')
ven_prod.head()
```

```python
#Read in our vendor dataset
vendor = pd.read_csv('vendors.csv')
vendor.head()
```

```python
vendor.info()
```

```python
#Merge the cve, union products and vendor datasets
union = pd.merge(cve, products, on = 'Unnamed: 0')
df = pd.merge(union, vendor, on = 'Unnamed: 0')
```

```python
#There are 15 features and 251380 entries
#13 of the features are categorical and we have 1 feature each of int and float
df.info()
```

```python
#We check for missing values in each feature in our dataset
df.isnull().sum()
```

```python
df.fillna(method='bfill', inplace = True)
```

```python
#We check the distribution of of our target data
plt.figure(figsize = (10, 7))
plt.hist(df['cvss'], bins = [x for x in range(11)], color = 'magenta')
plt.title('Distribution of CVSS scores', size = 16)
```

```python
df[pd.isnull(df['access_authentication']) != True]
```

```python
df['access_authentication'].value_counts()
```

```python
com_prod = df['vendor'].value_counts()
com_prod_cas = com_prod[com_prod >= 1000]
cal_list = [df[df['vendor']== x]['cvss'].mean() for x in com_prod_cas.index]
most_vulnerable_vendors = pd.Series(cal_list, index = com_prod_cas.index, name = 'Average CVSS')
most_vulnerable_vendors.sort_values(ascending = False)
```

```python
#We visualize our results
prod_vul = most_vulnerable_vendors.sort_values(ascending = False).head(10)
plt.figure(figsize = (10,7))
plt.title('Top Ten Most Popular Vendors With The Highest Average CVSS Score', size = 16)
plt.xlabel('Average CVSS Score', size = 12)
plt.ylabel('Vendors', size = 12)
prod_vul.plot(kind='barh', color = 'magenta')
```

```python
#Vendors with the most vulnerabilities
com_prod_max = df['vendor'].value_counts().head(10)

df_prod = pd.DataFrame({'Vendors': com_prod.index, 'No Of Cases Of Vulnerabilities': com_prod.values, 
                       '% of Total vulnerabilities': com_prod.values / len(df)})
df_prod
```

```python
plt.figure(figsize=(10,7))
plt.title('Top 10 Most Vulnerable Products', size = 16)
plt.xlabel('No Of Vunerabilities', size =12)
plt.ylabel('Products', size = 12)
df['vulnerable_product'].value_counts().head(10).plot(kind='barh', color = 'magenta')
```

```python
#The vendors outputted below has an average cvss score of 10
#There are 309 of such vendors
group_df = df.groupby('vendor').mean()['cvss'].sort_values(ascending=False)
group_df[group_df == 10]
```

```python
plt.figure(figsize = (10, 7))
plt.title('Most Common Products In Our Dataset', size = 16)
plt.xlabel('No Of Vulnerabilities', size = 12)
plt.ylabel('Vendors', size = 12)
df['vendor'].value_counts().head(10).plot(kind = 'barh', color = 'magenta')
```

```python
#Plots showing the average cvss score for each category in our categorical feature
list1 = ['access_authentication', 'access_complexity', 'access_vector', 
         'impact_confidentiality', 'impact_availability', 'impact_integrity']
fig, ax = plt.subplots(2,3)
fig.set_size_inches(16,10)
k = 0
for i in range(2):
    for j in range(3):
        ax[i][j].set_title(list1[k])
        cat = df.groupby(list1[k]).mean()['cvss']
        k = k + 1
        ax[i][j].bar(cat.index, cat.values, width =0.35, color = 'magenta')
fig.suptitle('DISTRIBUTION OF AVERAGE CVSS SCORE', size = 16)

        
```

```python
df['CVE YEAR'] = df['Unnamed: 0'].apply(lambda x: x.split('-')[1])
df['CVE YEAR'].head(10)
```

```python
#The average CVSS score for each year
df.groupby('CVE YEAR').mean()['cvss']
```

```python
df['mod_year'] = df['mod_date'].apply(lambda x: x.split(' ')[0])
df['mod_year'] = pd.to_datetime(df['mod_year'])
df_copy = df.copy()
df_copy = shuffle(df_copy, random_state = 0)
df_vis = df_copy.iloc[:200,:]
plt.figure(figsize = (12, 8))
plt.plot(df_vis['mod_year'], df_vis['cvss'], '--', marker = '*', color = 'magenta')

plt.xlabel('Year', size = 14)
plt.ylabel('CVSS', size = 14)
```

```python
df['pub_year'] = df['pub_date'].apply(lambda x: x.split(' ')[0])
df['pub_year'] = pd.to_datetime(df['pub_year'])
df_copy = df.copy()
df_copy = shuffle(df_copy, random_state = 0)
df_vis = df_copy.iloc[:200,:]
plt.figure(figsize = (12, 8))
plt.plot(df_vis['pub_year'], df_vis['cvss'], '--', marker = '*', color = 'magenta')

plt.xlabel('Year', size = 14)
plt.ylabel('CVSS', size = 14)
```

From our visualisations above there is no trend or seasonality in our data set, hence our dataset is stationary.
